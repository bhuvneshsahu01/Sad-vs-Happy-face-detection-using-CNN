```python
im=np.zeros((512,512,3))
def draw(event,x,y,flags,params):
    if event==0:
        cv2.circle(im,center=(x,y),thickness=2,radius=40,color=(5,67,200))
        cv2.rectangle(im,(x-40,y-40),(x+40,y+40),thickness=2,color=(100,200,34))
     
cv2.namedWindow(winname='im')
cv2.setMouseCallback('im',draw)
while True:
    cv2.imshow('im',im)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
cv2.destroyAllWindows()
```
